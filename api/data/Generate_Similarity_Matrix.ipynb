```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv("steam_games.csv")
```

```python
games.shape
```

```python
games.head(0)
```

```python
games.shape
```

```python
games["steam_appid"] = games.url.str.extract(r'(\d+(?=/))')
games["positive_review_percentage"] = games.all_reviews.str.extract(r'(\d+(?=%))')

games["review_count"] = games.all_reviews.str.replace(",", "")
games["review_count"] = games.review_count.str.extract(r'(\d+(?=))').replace(np.NaN, 0)
games["review_count"] = games.review_count.astype(int)
```

```python
games.dropna(inplace=True, subset=['game_description', 'name', 'genre', 'popular_tags', 'developer', 'publisher'])
games = games[games.review_count > 300]
```

```python
games = games[['steam_appid', 'name', 'game_description', 'genre', 'popular_tags', 'developer', 'publisher', 'positive_review_percentage', 'desc_snippet']]
```

```python
games.head()
```

```python
games.shape
```

```python
games.genre = games.genre.str.split(',')
games.popular_tags = games.popular_tags.str.split(',')
games.publisher = games.publisher.str.split(',')
games.developer = games.developer.str.split(',')
```

```python
games.genre, games.popular_tags
```

```python
games.game_description = games.game_description.str.split(' ').apply(lambda x: [i for i in x[4:] if i])
```

```python
games.game_description[266]
```

```python
def removeSpaces(col) : 
    games[col] = games[col].apply(lambda x: [i.replace(" ","") for i in x] if isinstance(x, list) else x)

for col in ['genre', 'popular_tags', 'publisher', 'developer']:
    removeSpaces(col)
```

```python
games = games.dropna(subset=['name'])
```

```python
games['tags'] = games['game_description'] + games['popular_tags'] + games['publisher'] + games['developer']
```

```python
games['tags']
```

```python
new_games = games[['steam_appid', 'name', 'tags']]
```

```python
new_games.head()
```

```python
new_games.tags = new_games.tags.apply(lambda x: " ".join(x) if isinstance(x, list) else x)
```

```python
new_games.tags = new_games.tags.str.lower()
```

```python
new_games.tags
```

```python
new_games = new_games.dropna(subset=['tags'])

print(new_games.isna().sum())
new_games = new_games.reset_index()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_games.tags).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    arr = []
    for i in text.split():
        a = ps.stem(i)
        arr.append(a)
    return " ".join(arr)
```

```python
new_games.tags = new_games.tags.apply(stem)
```

```python
new_games.tags[1]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
def reccomended(game):
    index = new_games[new_games.name == game].index[0]
    distances = similarity[index]
    similar_games = sorted( list(enumerate(distances)) ,reverse = True,key = lambda x : x[1])[1:11]
    print(new_games.shape)
    for i in similar_games:
        print(new_games.iloc[i[0]]['name'])
```

```python
reccomended("DARK SOULS™ III")
```
